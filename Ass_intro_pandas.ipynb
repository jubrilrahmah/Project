```python
#outer merge between custoer and product
cus_prod = pd.merge(customers, products, how = "outer" , left_on = "customer_id", right_on= "product_id")

cus_prod.head()
```

```python
#a left merge between orders and customers
ord_cus = pd.merge(orders, customers, how= "left", on ="customer_id")

ord_cus.head(20)
```

Using the left merger gives the result of inner merge(intersection) plus the right dataframe. from the code above, there is a common column between the two data frame (which is the customer _id), hence the return of the only right dataframe

```python
#right merge between product and customer
pro_cus = pd.merge(products, customers, how= "right", left_on= "product_id", right_on= "customer_id")

pro_cus.head()
```

Using the right merger gives the result of inner merge(intersection) plus the right dataframe. from the code above, there is no intersection(inner) between the two data frame, hence the return of the only right dataframe
